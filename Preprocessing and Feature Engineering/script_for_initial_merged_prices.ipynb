```python
# imports
import pandas as pd
import glob
import os
```

```python
# locate all CSV files
file_paths = glob.glob('full_history/*.csv')
print(f'Found {len(file_paths)} files:')
for p in file_paths:
    print('  ', os.path.basename(p))
```

```python
# read & tag each file by its symbol
df_list = []
for path in file_paths:
    symbol = os.path.splitext(os.path.basename(path))[0]
    df = pd.read_csv(path, parse_dates=['date'])
    df['symbol'] = symbol
    df_list.append(df)
```

```python
# merge into one DataFrame
merged = pd.concat(df_list, ignore_index=True)
```

```python
# sort by symbol and date
merged.sort_values(['symbol', 'date'], inplace=True)
```

```python
# reset index to get a clean 0...N-1 sequence
merged.reset_index(drop=True, inplace=True)
```

```python
# save
merged.to_csv('merged_full_history.csv', index=False)
print("Saved merged_full_history.csv with", len(merged), "rows")
```
